```python
import matplotlib.pyplot as plt
import numpy as np

hatch_patterns_class = {
    'UA': '///',  # Garis diagonal
    'PRE': 'xxx',  # Garis diagonal terbalik
}

# Data from the table
models = ['IndoBERT dengan Augmentasi', 'IndoBERT tanpa Augmentasi']
accuracy = [0.4359, 0.5128]
f1_score = [0.3654, 0.4785]
recall = [0.4127, 0.4960]
precision = [0.3669, 0.4951]


# Bar width
bar_width = 0.2
index = np.arange(len(models))

# Create bars with black and white color and hatch patterns
fig, ax = plt.subplots(figsize=(10, 6))

bar1 = ax.bar(index - bar_width*1.5, accuracy, bar_width, label='Accuracy', color='white', edgecolor='black', hatch='//')
bar2 = ax.bar(index - bar_width/2, recall, bar_width, label='Recall', color='white', edgecolor='black', hatch='\\')
bar3 = ax.bar(index + bar_width/2, precision, bar_width, label='Precision', color='white', edgecolor='black', hatch='--')
bar4 = ax.bar(index + bar_width*1.5, f1_score, bar_width, label='F1-Score', color='white', edgecolor='black', hatch='..')

# Labels and title
ax.set_xlabel('Model', fontsize=12)
ax.set_ylabel('Scores', fontsize=12)
ax.set_xticks(index)
ax.set_xticklabels(models)
ax.legend()

# Display the graph
plt.tight_layout()
plt.show()
```
